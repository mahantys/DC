```python
from numba import jit
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import os
```

#### Image 
An image is considered an array of intensity values .
So an __image__ of size 32x32 can be considered as an 2D array of size 32x32 for __grayscale__ image and 3D array of size 32x32x3 for a __R.G.B.__ image

Keras is used herefor reading/writing images but any other library like __PIL__ can be used

```python
img_path = "data/images"
img_name = "rose.jpg"
img = image.load_img(os.path.join(img_path, img_name), target_size=(176, 256), color_mode="grayscale")
```

### Displaying the loaded image

```python
orig_image = image.img_to_array(img)

# The image is in shape (176, 256, 1) . so reshaping it to (176 , 256)
orig_image = np.reshape(orig_image, (176, 256))
# Displaying the image
plt.imshow(orig_image, cmap="gray")
```

### Noise
Removing noise is generally the first step . One way of doing is averaging the value of the pixels surrounding the current pixel. This is generally done under the assumptions 

* Value of the pixel is similar to the nearby pixels
* Noise of the pixel is independent of the noise of the other pixels. This should average as zero for the combined picture
